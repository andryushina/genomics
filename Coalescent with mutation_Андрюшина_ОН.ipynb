```python
import numpy as np
import random
```

```python
def population_size(t, t_change, n_change):
    if len(t_change) == 1:
        return n_change[0]
    else:
        i = 0
        while i != len(t_change) - 1:
            if t_change[i] <= t and t < t_change[i + 1]:
                return n_change[i]
            i += 1
        if t_change[-1] <= t:
            return n_change[-1]
```

```python
def coal_with_mut(k, mu, t_change, n_change):
    t = 0
    muts = []
    coals = []
    while k != 1:
        l_m =  k * mu
        l_c = k * (k - 1) / 2 / population_size(t, t_change, n_change)
        t += np.random.exponential(1 / (l_c + l_m))
        event = np.random.binomial(1, l_c / (l_c + l_m))
        if event == 1:
            k1_c, k2_c = np.random.choice(k, size=2, replace=False)
            coal = [t, k1_c, k2_c]
            coals.append(coal)
            k -= 1
        else:
            k_m = np.random.choice(k)
            t_m = np.random.uniform()
            mut = [t, k_m, t_m]
            f = True
            if len(muts) != 0:
                for i in range(len(muts)):
                    if muts[i][1] == k_m:
                        f = False
            if f:
                muts.append(mut)
    return coals, muts
```

```python
k = 2
mu = 0
t = [0, 0.025, 0.325, 3]
n = [3, 0.1, 1.5, 3]
c, m = coal_with_mut(k, mu, t, n)
print('coalescence')
print('the first number - coalescence time, second and third numbers - pair of lineages')
print(c)
print('\n')
print('mutation')
print('the first number - mutation time, the second number - lineage, the third number -  mutation position on a genome')
print(m)
```
